```python
import pandas as pd
file_path = "/content/drive/MyDrive/Portfolio/Data_analysis_projects/ApacheKafka_Bug_Analysis/data/processed/kafka_bug_reports_cleaned.csv"
df = pd.read_csv(file_path)
```

```python
# Ensure components are treated as a string to allow splitting
df['components'] = df['components'].astype(str)

# Replace the "Unspecified Component" string so it's not split
df['components'] = df['components'].str.replace('Unspecified Component', 'UnspecifiedComponent_temp', regex=False)

# Split the components string into a list of components
df['components_split'] = df['components'].str.split(', ')

# Explode the DataFrame to create a new row for each component
df_normalized = df.explode('components_split').reset_index(drop=True)

# Clean up the exploded component names
df_normalized['components_split'] = df_normalized['components_split'].str.strip()

# Change the placeholder back to the original text
df_normalized['components_split'] = df_normalized['components_split'].str.replace('UnspecifiedComponent_temp', 'Unspecified Component', regex=False)

# You can now use the 'components_split' column for your Tableau chart
```

```python
# Save the new DataFrame to a new CSV in your processed folder
output_path = "/content/drive/MyDrive/Portfolio/Data_analysis_projects/ApacheKafka_Bug_Analysis/data/processed/kafka_bug_reports_normalized.csv"
df_normalized.to_csv(output_path, index=False)
print(f"Normalized data saved to {output_path}")
```
